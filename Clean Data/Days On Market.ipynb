```python
import pandas as pd
import numpy as np
```

```python
path1 = "Raw Data\MedianDaysToClose.csv"
path2 = "Raw Data\MedianDaysToPending.csv"
path3 = "Raw Data\ZillowHomeValues.csv"
path4 = "cleanpopulationdf.csv"
closedf = pd.read_csv(path1)
pendingdf = pd.read_csv(path2)
homevaldf = pd.read_csv(path3)
cleanpopulationdf = pd.read_csv(path4)


pendingdf.head()
```

```python
closedf.head()
```

```python
daysonmarketdf = pd.concat([closedf, pendingdf]).groupby(['RegionID', 'RegionName','SizeRank']).sum()
daysonmarketdf.head()
```

```python
count_row = daysonmarketdf.shape[0] 
count_row
```

```python
daysonmarketdf.to_csv('daysonmarket.csv')
```

```python
rows_to_drop = []
for i in daysonmarketdf.index.tolist(): #iterate over rows
    for j in daysonmarketdf.columns.tolist(): #iterate over columns
        value = daysonmarketdf.loc[i, j] #get cell value
        if (value == 0):
            rows_to_drop.append(i)
            break
len(rows_to_drop)
```

```python
rows_to_drop
```

```python
cleaned_df = daysonmarketdf.drop(index = rows_to_drop)
len(cleaned_df)
```

```python
cleanpopulationdf.head()
```

```python
combineddf = pd.merge(cleaned_df, cleanpopulationdf, how='inner', on='RegionName')
populationdaysfinaldf = combineddf.drop_duplicates(subset=["RegionName"], keep = 'first')
populationdaysfinaldf
```
